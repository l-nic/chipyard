```python
%matplotlib notebook
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
plt.rc('axes', labelsize=14)
plt.rc('legend', fontsize=12)
plt.rc('xtick', labelsize=12)
plt.rc('ytick', labelsize=12)
```

## Loopback Latency

```python
lat_path = 'results/latency/'
ndp_latency = pd.read_csv(lat_path + 'ndp_msg_len_latency.csv', index_col="msg_len").groupby('msg_len').median()/3.2
homa_latency = pd.read_csv(lat_path + 'homa_msg_len_latency.csv', index_col="msg_len").groupby('msg_len').median()/3.2
```

```python
plt.figure(figsize=(7, 3))
ax = plt.gca()
ndp_latency['latency'].plot(ax=ax, color='r', linestyle='--', label='NDP', marker='*', markersize=8)
homa_latency['latency'].plot(ax=ax, color='b', linestyle='--', label='Homa', marker='o', markersize=8)
ax.set_ylabel('Median Latency (ns)')
ax.set_xlabel('Message Length (B)')
ax.legend(loc='upper center', frameon=False, ncol=2, bbox_to_anchor=(0.5, 1.1))
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.yaxis.grid(alpha=0.6)
plt.tight_layout()

# plt.title('Loopback Latency')
plt.show()
```

## Loopback Throughput

```python
thp_path = 'results/throughput/'
ndp_tput = pd.read_csv(thp_path + 'ndp_msg_len_throughput.csv', index_col="msg_len")
homa_tput = pd.read_csv(thp_path + 'homa_msg_len_throughput.csv', index_col="msg_len")
```

```python
plt.figure(figsize=(7, 3))
ax = plt.gca()
ndp_tput['throughput'].plot(ax=ax, color='r', linestyle='--', label='NDP', marker='*', markersize=8)
homa_tput['throughput'].plot(ax=ax, color='b', linestyle='--', label='Homa', marker='o', markersize=8)
ax.set_ylabel('Throughput (Gbps)')
ax.set_xlabel('Message Length (B)')
ax.legend(loc='upper center', frameon=False, ncol=2, bbox_to_anchor=(0.5, 1.1))
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.yaxis.grid(alpha=0.6)
plt.tight_layout()

# plt.title('Loopback Throughput Speedup')
plt.show()
```
